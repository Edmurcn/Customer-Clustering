- Nesta etapa estudamos a performance do modelo ao realizar a clusterização. Desse modo, vamos avaliar se o modelo nos trouxe informação suficiente para respondermos ao problema de negócio proposto.

```python
# importar as bibliotecas

import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# importando os dados com a label dos clusters

df = pd.read_csv("/home/edmurcn/Documentos/MeusProjetos/Customer-Clustering/data/dados_label.csv", header=0)
```

```python
# Análise da clusterização

palette = sns.color_palette("flare")

#Gráfico do tamanho dos clusters

plt.figure(figsize=(6,4), facecolor='lightgray')
sns.countplot(df, x="Label", orient="h", palette=palette)
plt.xlabel("Grupos", fontsize=13)
plt.ylabel("Quantidade", fontsize=13)
plt.show()

fig, ax = plt.subplots(3, 3, figsize=(16,10), facecolor="lightgray")

def col_row(n):
    col = n % 3
    row = n // 3
    return row, col

# loop para a criação dos gráficos das distribuições das colunas diferenciado pelo cluster

for i in range(0, df.shape[1]-1):

    row, col = col_row(i)
    coluna = df.columns[i]
    sns.kdeplot(data=df, x=coluna, hue="Label", ax=ax[row,col])
    ax[row, col].set_xlabel(coluna)
    ax[row, col].grid(True, linestyle='--', c="lightgray" )

```

#### Resolução do Problema de Negócio

- Para elencar as propostas com base nas características de cada cluster, levamos em conta a compatibilidade de cada grupo entre clientes leais e clientes potenciais, e para atingir cada cliente vamos colocar em prática o marketing *one-to-one*, ou seja, tornar o tratamento com cada cliente personalizado, buscando vislumbrar as expectativas individuais. Dessa forma, construímos as seguintes propostas:

    - **Plano Conforto**: Em busca de atingir grupos mais idosos, será proposto a implementação de um sistema de entregas de mercadoria própio do mercado, onde os clientes encomendam suas compras e agendam seus horários de entrega, essa prática faz com que o estabelecimento mantenha contato direto com o cliente por redes sociais contribuindo para que haja o aumento da frequência de compras, além de ser uma oportunidade para reter dados a respeito de horários e preferência de produtos e com isso a equipe de marketing poderia interagir com ofertas e campanhas personalizadas.
  
        O insight para a apropriação desse plano surge ao analisarmos a idade dos clientes e o potencial de compra, além disso, podemos atingir novos indivíduos que se enquadram neste grupo, os quais por conta de locomoção ou hábito ainda não são clientes do mercado.

    - **Plano Agilidade**: Pensando na rotina do grupo de homens que residem em grandes cidades e estão em cargos mais importantes com tarefas a todo momento e com pouco tempo livre para se alimentar, propomos uma estratégia para se adequar a esse situação, o plano consiste em instituir pequenos pontos 24 horas em grandes centros, nestes pontos estariam presentes somente alimentos prontos para consumo rápido e que seriam ágeis e de fácil consumo. Aproveitando a mudança de cultura da empresa ao abordar essa estratégia, talvez um algo a mais não seria tão agressivo, como a implantação de ponto de alimetação também 24 horas, onde os clientes poderiam utilizar de microondas para descongelar alimentos fornecidos pelo estabelecimento. 
    
        Busca-se com essa abordagem atender as necessidades desse grupo, que já são clientes numerosos e tem grande potencial de compra.

    - Pensando nos outros grupos de cliente, temos um grupo numeroso mas com um perfil adverso para personalização, a equipe de marketing poderia divulgar ofertas e campanhas para esses grupos, poreḿ essa prática não teria relação com a clusterização, apenas uma prática racional para mercados. Outro ponto que contribui para ssa decisão é o fato de que esses indivíduos já são leais ao estabelecimento e em análise das características não encontramos um grande potencial para eles.

- Desse modo, chegamos nas estratégias de mercado por meio da aplicação de ferramentas de aprendizado de máquina, esta integração revoluciona os setores de negócios e marketing dos estabelecimentos, trazendo benefícios para os dois lados tanto empresa quanto clientes.
